##### MODE IMPUTATION 
#### Impute Value  :
1) AMT_GOODS_PRICE  == > 
For consumer loans it is the price of the goods for which the loan is given,this loan varies based on the product and 
hence we cant fill with mean,median etc, hence i suggested to use NULL for those missing values.
since its very minimal missing values doesnt going to affect the analysis.

2) CNT_FAM_MEMBERS == > 
As per my analysis on this particular column, max number of applicant having 2 members in their family,
by average (~2 ) we can fill for the missing values.

3) EXT_SOURCE_2 == > 
This is normalized score from external data source, so we can fill with average (avg/mean/median) values to all 
those missing values.

4) OBS_30_CNT_SOCIAL_CIRCLE,DEF_30_CNT_SOCIAL_CIRCLE,OBS_60_CNT_SOCIAL_CIRCLE,DEF_60_CNT_SOCIAL_CIRCLE  == > 
either 0 or blank can use to fill those missing values.

5)AMT_ANNUITY  == >
Based on the analysis we can infer that we need separate formulae to calculate the Amt_annuity- 
hence avg/mean/median etc method wont applicable, hence we can use NULL to fill all those missing values

```python
### Treating the outliers just by filtering the values based on the computed percentile

P = np.percentile(appli_Data_df.AMT_INCOME_TOTAL, [5, 95])
new_df = appli_Data_df[(appli_Data_df.AMT_INCOME_TOTAL > P[0]) & (appli_Data_df.AMT_INCOME_TOTAL < P[1])]
```
